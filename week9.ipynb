```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
from skimage import io
from skimage import transform
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.morphology import (erosion, dilation, closing, opening,
                                area_closing, area_opening)
from skimage.color import rgb2gray
```

```python
img = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/j.png')

kernel = np.ones((5,5), np.uint8)
dilation = cv.dilate(img, kernel, iterations = 1)

plt.subplot(121),plt.imshow(img),plt.title('Citra Awal')
plt.subplot(122),plt.imshow(dilation),plt.title('Citra Hasil Dilasi')
```

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv.morphologyEx(img, cv.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initialize Citra')
plt.subplot(122), plt.imshow(erosion), plt.title('Results from Eruption Citra')
```

```python
kernel = np.ones((5,5), np.uint8)
dilation = cv.morphologyEx(img, cv.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initialize Citra')
plt.subplot(122), plt.imshow(dilation), plt.title('Results from Dilation Citra')
```

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv.morphologyEx(img, cv.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Initialize Citra')
plt.subplot(122), plt.imshow(erosion), plt.title('Results from Eruption Citra')
```

```python
#Without Library 
kernel = np.ones((7,7), np.uint8)
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(opening), plt.title('Citra Hasil Opening')
```

```python
#With Library
element = cv.getStructuringElement(cv.MORPH_RECT,(7,7))
erosion_dst = cv.erode(img, element)
opening_dst = cv.dilate(erosion_dst, element)

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(opening_dst), plt.title('Citra Hasil Opening')
```

```python
#Without Library
kernel = np.ones((7,7), np.uint8)
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(closing), plt.title('Citra Hasil Closing')
```

```python
#With Library
element = cv.getStructuringElement(cv.MORPH_RECT,(7,7))
dilate_dst = cv.dilate(img, element)
closing_dst = cv.erode(dilate_dst, element)

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(closing_dst), plt.title('Citra Hasil Closing')
```

```python
#(3X3)
kernel = cv.getStructuringElement(cv.MORPH_CROSS, (3, 3))
cross3 = cv.morphologyEx(img, cv.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(cross3), plt.title('Citra Hasil Dilasi 3x3')
```

```python
#(5X5)
kernel = cv.getStructuringElement(cv.MORPH_CROSS, (5, 5))
cross3 = cv.morphologyEx(img, cv.MORPH_DILATE, kernel)
plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(cross3), plt.title('Citra Hasil Dilasi 5x5')
```
