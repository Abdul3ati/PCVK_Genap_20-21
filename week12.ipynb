```python
cascade_wajah = cv.CascadeClassifier('/content/drive/MyDrive/PCVK_MM/haarcascades/haarcascade_frontalcatface_extended.xml')
kucing = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/kucing.jpg')
kucing_gray = cv.cvtColor(kucing, cv.COLOR_BGR2GRAY)
kucing_gray = cv.cvtColor(kucing_gray, cv.COLOR_GRAY2RGB)
roi_wajah = cascade_wajah.detectMultiScale(kucing_gray, scaleFactor=1.05, minNeighbors=5)
for(x,y,w,h) in roi_wajah:
  # print(x)
  if (x<400):
    cv.rectangle(kucing_gray,(x,y),(x+w,y+h),(0,255,0),3)
plt.figure(figsize = (7,7))
plt.imshow(kucing_gray, cmap='gray')
```
